# Parte 2 - Criando o Modelo

```python
#%matplotlib inline (para o jupyter)
from sklearn.datasets import load_iris
from pandas.plotting import scatter_matrix
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
dados_iris = load_iris()
```

```python
X_train, X_test, y_train, y_test = train_test_split(dados_iris['data'], dados_iris['target'], random_state=0)
```

75% Treino, 25% Teste

```python
#150 observaes
print(f'Treino: {X_train.shape}')
print(f'Teste: {X_test.shape}')
```

```python
iris_dataframe = pd.DataFrame(X_train, columns= dados_iris.feature_names)
```

```python
grr = scatter_matrix(iris_dataframe, c= y_train, figsize= (15, 15), marker= 'o', hist_kwds= {'bins': 20}, s= 60, alpha= 0.8)
```

```python
knn = KNeighborsClassifier(n_neighbors= 1)
```

```python
knn.fit(X_train, y_train)
```

```python
print(f'Score Treino: {knn.score(X_train, y_train)}')
print(f'Score Teste: {knn.score(X_test, y_test)}')
```

```python
florx = np.array([[6.3, 4., 6, 2]])
```

```python
print(knn.predict(florx))
```

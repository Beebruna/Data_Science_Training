Retira a Setosa, por ser bastante diferente das outras

```python
train = train.query("Species == 'Iris-versicolor' | Species == 'Iris-virginica'")
```

```python
test = test.query("Species == 'Iris-versicolor' | Species == 'Iris-virginica'")
```

```python
train
```

```python
train['Species'] = np.where(train['Species'] == 'Iris-versicolor', 0, 1)
```

```python
test['Species'] = np.where(test['Species'] == 'Iris-versicolor', 0, 1)
```

```python
train_x = train.drop('Species', axis = 1)
```

```python
train_y = train['Species']
```

```python
test_x = test.drop('Species', axis = 1)
```

```python
test_y = test['Species']
```

```python
test_x
```

Treinar = Fit

```python
LR = LogisticRegression().fit(train_x, train_y)
```

```python
LIR = LinearRegression().fit(train_x, train_y)
```

```python
LDA = LinearDiscriminantAnalysis().fit(train_x, train_y)
```

```python
DT = DecisionTreeClassifier().fit(train_x, train_y)
```

```python
KNN = KNeighborsClassifier().fit(train_x, train_y)
```

```python
SVM = SVC().fit(train_x, train_y)
```

```python
RF = RandomForestClassifier().fit(train_x, train_y)
```

Verificando o melhor modelo

```python
print('Acurácia LR:', LR.score(test_x, test_y))
print('Acurácia LIR:', LIR.score(test_x, test_y))
print('Acurácia LDA:', LDA.score(test_x, test_y))
print('Acurácia DT:', DT.score(test_x, test_y))
print('Acurácia KNN:', KNN.score(test_x, test_y))
print('Acurácia SVM:', SVM.score(test_x, test_y))
print('Acurácia RF:', RF.score(test_x, test_y))
```

Matriz de Confusão + Métricas

image.png

```python
plot_confusion_matrix(LR, test_x, test_y)
plt.show()
```

image.png

```python
plot_roc_curve(LR, test_x, test_y, ax=plt.gca())
plot_roc_curve(LDA, test_x, test_y, ax=plt.gca())
plot_roc_curve(DT, test_x, test_y, ax=plt.gca())
plot_roc_curve(KNN, test_x, test_y, ax=plt.gca())
plot_roc_curve(SVM, test_x, test_y, ax=plt.gca())
plot_roc_curve(RF, test_x, test_y, ax=plt.gca())
```
